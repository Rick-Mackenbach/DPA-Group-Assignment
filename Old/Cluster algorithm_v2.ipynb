```python
import collections
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from pprint import pprint
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
sentences = []

with open('descriptions.txt', encoding = "utf8") as f:
    for line in f:
        text = line.lower()                                       ## Lowercase all characters
        text = text.replace("[comma]"," ")                        ## Replace [commas] with empty space
        for ch in text:
            if ch < "0" or (ch < "a" and ch > "9") or ch > "z":   ## The cleaning operation happens here, remove all special characters
                text = text.replace(ch," ")
        text = ' '.join(text.split())                             ## Remove double spacing from sentences
        sentences.append(text)
        
#sentences = sentences[0:100]
nclusters = int(len(sentences)/5)

print(len(sentences), nclusters)
```

```python
def word_tokenizer(text):
            #tokenizes and stems the text
    tokens = word_tokenize(text)
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(t) for t in tokens if t not in stopwords.words('english')]
    return tokens
```

#### Changed the code to also output the fit_predict array which is a 2D array where the index corresponds to the line of the text and the value at that index corresponds to the cluster value
previously looked at the dictionary type and replicating each dictionary index value for the number of items in the cluster but then would have duplicates

```python
def cluster_sentences(sentences, nb_of_clusters):
    tfidf_vectorizer = TfidfVectorizer(tokenizer=word_tokenizer,
                                    stop_words=stopwords.words('english'),
                                     max_df=0.9,
                                    min_df=0.1,
                                    lowercase=True)
            #builds a tf-idf matrix for the sentences
    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
    kmeans = KMeans(n_clusters=nb_of_clusters)
    kmeans.fit(tfidf_matrix)
    
    kmeans_array = kmeans.fit_predict(tfidf_matrix)
    
    clusters = collections.defaultdict(list)
    for i, label in enumerate(kmeans.labels_):
            clusters[label].append(i)


    return dict(clusters), kmeans_array
```

```python
clusters, kmeans_predict = cluster_sentences(sentences, nclusters)
```

For every line in the fit_predict array, it finds those lines that belong to the same cluster and adds them to a list. Then it is finished it saves the list to a bigger list.

```python
output = []

for i in range(len(kmeans_predict)):
    temp = [i]
    for j in range(1,len(kmeans_predict)):
        jl = [j]
        if kmeans_predict[i]==kmeans_predict[j] and j not in temp:
            temp += [j]
    output.append(temp)
```

Then for lists with more than 5, we remove the last values. For lists with less than 5 we duplicate the second value

```python
for line in output:
    if len(line) > 5:
        index = output.index(line)
        output[index] = line[0:5]
    if len(line) > 1  and len(line) < 5:
        line += [line[1]]*(5-len(line))
    if len(line) == 1:
        line += [line[0]]*4

for line in output:
    print(line)
```

Import CSV module write each row into the file. Currently using a true comma separator.

```python
import csv

with open('results.csv', 'w', newline='') as csvfile:
    results = csv.writer(csvfile, delimiter=',')
    for line in output:
        results.writerow(line)
```
